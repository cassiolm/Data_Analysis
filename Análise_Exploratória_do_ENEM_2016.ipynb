## Análise de nota da Redação por Sexo

```
colunas_selecionadas = ['TP_SEXO', 'NU_NOTA_REDACAO']
```

```
microdadosEnemSexoRedacao = microdadosEnem.filter(items=colunas_selecionadas)
```

```
microdadosEnemSexoRedacao.head()
```

##**Remoção de NA's**

```
microdadosEnemSexoRedacao = microdadosEnemSexoRedacao.dropna()
```

```
microdadosEnemSexoRedacao.head()
```

## **Contagem de notas por sexo**

```
microdadosEnemSexoRedacao.groupby('TP_SEXO').count()
```

## **Nota máxima e mínima são as mesmas por sexo?**

```
# Sexo feminino aqui tem a nota máxima maior que o masculino
microdadosEnemSexoRedacao.groupby('TP_SEXO').max()
```

```
microdadosEnemSexoRedacao.groupby('TP_SEXO').min()
```

```
## Aqui vamos eliminar as notas menores que zero
# Sexo feminino aqui tem a nota mínima maior que o masculino
microdadosEnemSexoRedacao[microdadosEnemSexoRedacao.NU_NOTA_REDACAO >0].groupby('TP_SEXO').min()
```

##**Métodos estatísticas para verificar o sexo com a melhor nota de redação**

```
#Média simples
microdadosEnemSexoRedacao.groupby('TP_SEXO').mean()
```

```
#Mediana
microdadosEnemSexoRedacao.groupby('TP_SEXO').median()
```

```
#Histograma para verificar a distribuição, que é muito parecida inclusive.
microdadosEnemSexoRedacao.groupby('TP_SEXO').hist()
```

```
## Resumindo todas as métricas
microdadosEnemSexoRedacao.groupby('TP_SEXO').describe()
```

**Conclusão retirada do conjunto de dados explorado é que, dentro das variáveis exploradas, o sexo feminino tem uma vantagem na nota de redação com relação ao sexo masculino.**

# **Passo 1 - Criando um dicionário:**

```
q001e002Dicionário = {'A': 'Nunca estudou', 
                      'B': "Não completou a 4ª série/5º ano do Ensino Fundamental.",
                      'C': 'Completou a 4ª série/5º ano, mas não completou a 8ª série/9º ano do Ensino Fundamental',
                      "D": "Completou a 8ª série/9º ano do Ensino Fundamental, mas não completou o Ensino Médio.",
                      "E": "Completou o Ensino Médio, mas não completou a Faculdade",
                      "F": "Completou a Faculdade, mas não completou a Pós-graduação",
                      "G": "Completou a Pós-graduação",
                      "H": "Não sei.", 
                      }
```

```
microdadosEnemSelecionado.filter(items=['Q001','NU_INSCRICAO']).groupby('Q001').count()
```

#**Passo 2 - Criar novas colunas no DF**

###Inserido coluna NO_Q001

```
microdadosEnemSelecionado['NO_Q001'] = [ q001e002Dicionário[resp] for resp in microdadosEnemSelecionado.Q001]
```

```
microdadosEnemSelecionado.head()
```

###Inserido coluna NO_Q002

```
microdadosEnemSelecionado['NO_Q002'] = [ q001e002Dicionário[resp] for resp in microdadosEnemSelecionado.Q002]
```

###Verificação do Head para checar a inserção

```
microdadosEnemSelecionado.head()
```

#**Passo 3 - Comparação da distribuição da escolaridade de Pais e Mães**

Hora de utilizar os métodos do DataFrame como filter(), groupby, count(), sort_values()

1 - Distribuição dos candidatos do Enem 2016 por escolaridade do pai

```
microdadosEnemSelecionado.filter(items=['NU_INSCRICAO', 'NO_Q001']).groupby('NO_Q001').count().sort_values(by='NU_INSCRICAO', ascending = False)
```

2 - Distribuição dos candidatos do Enem 2016 por escolaridade da mãe

```
microdadosEnemSelecionado.filter(items=['NU_INSCRICAO', 'NO_Q002']).groupby('NO_Q002').count().sort_values(by='NU_INSCRICAO', ascending = False)
```

##**Curiosidade - Desempenho em matemática segundo a escolaridade do Pai e Mãe (Correlação?)**

1 - por escolaridade do pai

```
microdadosEnemSelecionado.filter(items=['NU_NOTA_MT', 'NO_Q001']).groupby('NO_Q001').mean().sort_values(by='NU_NOTA_MT', ascending = False)
```

2 - por escolaridade da mãe

```
microdadosEnemSelecionado.filter(items=['NU_NOTA_MT', 'NO_Q002']).groupby('NO_Q002').mean().sort_values(by='NU_NOTA_MT', ascending = False)
```

Podemos inferir que quanto maior a escolaridade dos pais dos candidatos, maior sua nota em matemática (indicativo de correlação devido a linearidade das respostas) mas será que esta tendência também se reflete em outras notas?

##**Passo 5 - Olhando o desempenho em redação segundo a escolaridade dos pais**

1 - por escolaridade do pai

```
microdadosEnemSelecionado.filter(items=['NU_NOTA_REDACAO', 'NO_Q001']).groupby('NO_Q001').mean().sort_values(by='NU_NOTA_REDACAO', ascending = False)
```

2 - por escolaridade da mãe

```
microdadosEnemSelecionado.filter(items=['NU_NOTA_REDACAO', 'NO_Q002']).groupby('NO_Q002').mean().sort_values(by='NU_NOTA_REDACAO', ascending = False)
```

### **É possível inferir que existe uma correlação entre as notas dos candidatos e a escolaridade dos pais, tanto para notas de matemática quanto de redação**
